```python
import numpy as np
import math
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
a=0.01
b=0.02
N=11400
I=250
R=47
S=N-I-R
```

```python
x0=np.array([S,I,R])
t=np.arange(0,200,0.01)
```

```python
def syst1 (x,t):
    dx_1=0
    dx_2=-b*x[1]
    dx_3=b*x[1]
    return [dx_1, dx_2, dx_3]
```

```python
y1=odeint(syst1, x0, t)
```

```python
plt.plot(t, y1, lw=2)
plt.title('Динамика изменения числа людей в каждой из трех групп в случае 1')
plt.grid('True')
```

```python
def syst2 (x,t):
    dx_1=-a*x[0]
    dx_2=a*x[0]-b*x[1]
    dx_3=b*x[1]
    return [dx_1, dx_2, dx_3]
```

```python
y2=odeint(syst2, x0, t)
```

```python
plt.plot(t, y2, lw=2)
plt.title('Динамика изменения числа людей в каждой из трех групп в случае 2')
plt.grid('True')
```
